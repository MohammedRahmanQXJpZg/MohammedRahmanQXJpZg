## Q1: Candidate Number


**15 pts**

In the box below, set the variable `CN` to your candidate number. Then in the next box write *a function* that takes an integer input and computes the sum of digits of the integer, divides this sum by 3 and returns the remainder (i.e., returns the sum of the digits <a href="https://en.wikipedia.org/wiki/Modulo_operation" target="_blank">mod 3</a>). 

For example, `1234` would return `1` since `1+2+3+4 = 10` and `10%3 == 1` 

Call this function `cnm3`.

**HINT** A valid but efficient method would be to convert the input integer to a string and loop over (and sum) the characters of the string. A faster method is to keep the input an integer and repeatedly divide by 10, taking the ones digit each time. Towards  this, the following operations might be helpful: `//` is integer division, `n // d` divides `n` by `d` *ignoring* and remainder. `%` is, of course, the mod operation that produces the remainder.

```python
#enter your (actual) candidate number
CN = 2107622
```

```python
#complete function cnm3
def cnm3(n):
    """
    input: n is an integer
    output: integer in {0,1,2}, sum of digits mod 3
    """
    #Function takes an integer input, computes the sum of all digits and then takes mod 3 of that sum.
    n_str=str(n) #Computes the input in string form
    n_sum=0 #Helper variable used to count the sum of n
    for i in n_str: #Iterates ith times in the string form of n
        n_sum += int(i) #Appends the ith integer of n onto the helper variable creating a summation of n
    n_mod=n_sum%3 #Computes the mod 3 of the sum
    return n_mod #Returns the mod 3 as an integer in {0,1,2}
```

## Q2: Random Groups

Write a function, called `rand_groups` which takes as an input a list of student numbers and generates as output *random* groups of `n` students, exhausting the list. The number `n` should equal your key plus 3. For example, if your key was 1 (i.e., if `cnm3(CN) = 1`) then you should create random groups of `n = 3 + 1 = 4` students. If there is one group with less students that is okay. 

The output should be a dictionary with keys as integers, starting at 0, and values the corresponding groups (the groups can be lists of students). 

**HINT** Use a nested loop. On the outside, a `while` loop that keeps looping until we run out of students, on the inside loop, a `for` loop, draw key + 3 random students, add them to a list, remove them from the master list of students, then add this list of key + 3 students to the output. 

The random package (import below) contains the method
`random.randint(low,high)` to draw a random integer between `low` and `high` (both included). So to draw a single random student for a list called `lst` requires `random.randint(0,len(lst)-1)`.

```python
#create function rand_groups
def rand_groups(l):
    """
    input: l is a list of integers
    output: final_dict is a Dictionary which returns Integers:Groups (respective to Keys:Values)
    """
    #Function inputs a list of integers and divides them into random groups of size n and generates a dictionary output
    n=cnm3(CN)+3 #Computes variable n
    rand_l=[] #Creates empty list to randomize master list
    for i in range(len(l)-1): #Iterates the loop ith times in the range of the length of l-1 
        i=np.random.randint(0, len(l)-1) #Generates a random number from 1 to the length of l-1 (represents the postion)
        rand_l.append(l[i]) #Appends the random value assigned to the position to the empty list
        l.remove(l[i]) #Removes the value from the old list 
    if len(l)==1: #Tests if the old list is of size 1 (Eventually it will be as the loop does not remove all values)
        j=np.random.randint(0, len(rand_l)-1) #Generates j, a random number from 1 to the length of rand_l-1 
        rand_l.insert(j, l[len(l)-1]) #Inserts the final value of l into the (random) jth postion of rand_l
        l.remove(l[len(l)-1]) #Removes the final value from l
    final_l=[rand_l[k:k+n] for k in range(0,len(rand_l),n)] #Using list comprehension, creates final list consisting of sublist of size n (adapted from geeksforgeeks.org)
    for m in range(len(final_l)): #Iterates mth times in range of the length of final_l
        final_dict={m:final_l[m]} #Generates a dictionary where each value of m is assigned to the mth value in final_l
        print(final_dict) #prints the entire dictionary
    return
```

## Q3: Three Gamblers


**30 pts**


Three brothers Dmitri, Ivan, and Alexei are playing a gambling game. They roll a three sided die with three equally likely outcomes: $\{0,1,2\}$. 

 - If a 0 is rolled: Dimitri pays each of his brothers \$k where `k` is the value of your key above (i.e., either 0, 1, or 2 depending on your CN in Q1). That is, he looses 2k dollars and each other brother gains k dollars.
 - If a 1 is rolled: Ivan pays Dimitri 1 dollar (so D gains a dollar, I loses a dollar and A retains his wealth).
 - If a 2 is rolled: Alexei pays Dimitri 1 dollar (so D gains a dollar, A loses a dollar and I retains his wealth).
 
The brothers play 1000 times. Simulate this and store the results in a DataFrame where each column is a brother and the nth row contains the brothers wealth after n rolls. 

Add two column storing the outcome of the roll and the current leader (who has the highest wealth).

**Hint** this is possible using a for loop, and drawing a random integer each loop, calculating the change in wealth and then appending the row onto the DataFrame. It is also possible (and likely less coding) to calculate all the rolls as once using a Numpy array, similar to the Random Walk example at the end of Lecture 3. 

```python
#create dataFrame 
k=cnm3(CN) #Assigning key
D='Dimitri'  #Defines Players
I='Ivan'
A='Alexei'
payoff_dict={0:{D:-2*k , I:1*k, A:1*k}, 1:{D:1, I:-1, A:0}, 2:{D:1, I:0, A:-1}} #Defines payoff Dictionary for all players
n=0 #Helper variable for round counting
round_list=[] #Empty list for recording rounds
roll_list=[] #Empty list for recording rolls for each round
D_count=0 #Helper variable for accumulating payoffs for each player
I_count=0
A_count=0
D_list=[] #Empty list to record payoffs for each player per round
I_list=[]
A_list=[]
Leader_list=[] #Empty list to record leader for each round 

for n in range(1000): #Iterates n until desired round limit where each iteration simulates 1 round played
    n+=1 #Increase n by 1 every iteration
    r=np.random.randint(0, 3) #Assigns a random number in the range to r for that round
    D_count +=payoff_dict[r][D] #Appends the payoff to the counter per round for each player using the payoff dictionary 
    I_count +=payoff_dict[r][I]
    A_count +=payoff_dict[r][A]
    D_list.append(D_count) #Records the payoff by appending to the list per round for each player
    I_list.append(I_count)
    A_list.append(A_count)
    
    #Draw Logic:
    if max(D_count,I_count,A_count) == D_count and  D_count == I_count: #Tests if the leader's wealth is equal to the others
        Leader_list.append('Tie') #Records the wealth leader as a tie by appending to the list
    elif max(D_count,I_count,A_count) == D_count and  D_count == A_count:
        Leader_list.append('Tie')
    elif max(D_count,I_count,A_count) == I_count and I_count==A_count:
        Leader_list.append('Tie')
        
    if D_count > I_count and D_count > A_count: #Tests if the wealth of Dimitri is greater than Ivan and ALexei 
        Leader_list.append(D) #Records the wealth leader as Dimitri by appending to the list
    elif I_count > D_count and I_count > A_count: #Tests if the wealth of Ivan is greater than Dimitri and ALexei 
        Leader_list.append(I) #Records the wealth leader as Ivan by appending to the list
    elif A_count > D_count and A_count > I_count: #Tests if the wealth of Alexei is greater than Ivan and Dimitri 
        Leader_list.append(A) #Records the wealth leader as Alexei by appending to the list
        
    round_list.append(n) #Records the number of rounds as a list by appending all values of n
    roll_list.append(r) #Records every roll that took place as a list by appending all values of r
    
game_dict={D:D_list, I:I_list, A:A_list, 'Roll':roll_list, 'Leader':Leader_list} #Creates a Dictionary with the wealth of each player,  the roll list and the list of leaders

df=pd.DataFrame(game_dict, columns=[D, I, A, 'Roll', 'Leader'], index=round_list) #Generates a DataFrame using the dictionary with the specified index and columns
df
```
