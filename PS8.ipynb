Create a function, `pd_from_sym_list`, similar to the function from PS5, `pd_from_sym`, but that takes a list of symbols. This will take a list of strings as an input (each string should be a stock symbol) and returning a data frame with the columns open, close, movement, and two additional columns **symbol** and **open_perc**. `symbol` is just symbol assocated with the observation. `open_perc` is the perecentage change since the first entry. That is, the earliest date for each symbol should have a `open_perc` of 1 (or 100, depending on taste) and increases or decreases according to the precentage change over time (if it goes up 2% in the first day, the second date should be 1.02).  If the list argument has only 1 entry then the returned dataframe should be the same as from last week (except with the extra rows). If there are multiple symbols then it should return the concatination of each individual dataframe.  

```python
def pd_from_sym_list(sym_list):
    """
    input: A list of stock symbols
    output: A Dataframe with 5 columns for each stock in sym_list sorted by day
    A function that uses an API, generates a DataFrame with columns open, close, movement, symbol and open_perc  
    """
    response_list=[] #Empty list to track responses
    frame_list=[] #Empty list to concatonate DataFrames
    for i in range(len(sym_list)):
        api='https://www.alphavantage.co/query?' #Obtains API data
        data={"function":"time_series_daily","symbol":sym_list[i],"outputsize":"full","apikey":"ET9DLN821TXY76P8",}
        response_list.append(requests.get(api, params=data)) #Stores response for each stock
    for n in range(len(response_list)): #Iterates for Every Response
        ticker_dict=json.loads(response_list[n].content)
        frame=pd.DataFrame(ticker_dict['Time Series (Daily)']).T 
        frame.columns=list(map(lambda x: x[3:], frame.columns))
        frame.open=frame.open.astype(float) #Open Column
        frame.close=frame.close.astype(float) #Close Column
        frame['movement']=frame['close']-frame['open'] #Movement Column
        frame['symbol']=sym_list[n] #Symbol Column
        frame=frame.drop(columns=['high','low','volume']) #Drops unecessary columns 
        frame=frame.sort_index() #Sorts DataFrame by day
        frame['open perc']=frame.groupby('symbol')['movement'].apply(lambda x: x.div(x.iloc[0]).mul(100)) #Open Perc Column by Stock
        frame_list.append(frame) #Stores DataFrame for each stock
    df=pd.concat(frame_list) #Concatonates list of DataFrames into a single DataFrame
    df=df.sort_index() #Sorts Entire DataFrame by day (again)
    return df
```
